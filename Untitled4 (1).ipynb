```python
!pip install groq

from groq import Groq
import time

# Paste your Groq API key here
API_KEY = "your_actual_key_here"

client = Groq(api_key=API_KEY)

def ask_llm(prompt, temperature=0.7, model="llama-3.1-8b-instant"):
    response = client.chat.completions.create(
        model=model,
        messages=[{"role": "user", "content": prompt}],
        temperature=temperature
    )
    return response.choices[0].message.content

print("Setup complete! Groq + LLaMA ready!")
```

```python
# EXPERIMENT 1: How prompt structure affects LLM output

print("EXPERIMENT 1: PROMPT STRUCTURE ANALYSIS")
print("=" * 60)

prompts = {
    "Vague Prompt": "Tell me about climate change.",

    "Structured Prompt": """You are an environmental scientist.
Explain climate change in exactly 3 points:
1. Cause
2. Effect
3. Solution
Keep each point under 2 sentences.""",

    "Few Shot Prompt": """Here are examples of good summaries:
Example 1: 'Photosynthesis converts sunlight into energy using chlorophyll.'
Example 2: 'Gravity pulls objects toward Earth's center at 9.8 m/s².'

Now summarize climate change in one sentence like the examples above:"""
}

for prompt_type, prompt in prompts.items():
    print(f"\n PROMPT TYPE: {prompt_type}")
    print("-" * 40)
    response = ask_llm(prompt)
    print(response)
    print("=" * 60)
    time.sleep(2)

print("\nExperiment 1 Complete!")
```

```python
# EXPERIMENT 2: How temperature affects LLM output

print("EXPERIMENT 2: TEMPERATURE ANALYSIS")
print("=" * 60)

test_prompt = "Write a one paragraph story about a robot learning to paint."

temperatures = {
    "Low Temperature (0.1) - More Focused": 0.1,
    "Medium Temperature (0.7) - Balanced": 0.7,
    "High Temperature (1.5) - More Creative": 1.5
}

for temp_name, temp_value in temperatures.items():
    print(f"\n {temp_name}")
    print("-" * 40)
    response = ask_llm(test_prompt, temperature=temp_value)
    print(response)
    print("=" * 60)
    time.sleep(2)

print("\nExperiment 2 Complete!")
```

```python
# EXPERIMENT 3: How context length affects LLM output

print("EXPERIMENT 3: CONTEXT LENGTH ANALYSIS")
print("=" * 60)

prompts = {
    "No Context": "What is photosynthesis?",

    "Medium Context": """Photosynthesis is a biological process.
Plants use sunlight to make food.
Chlorophyll is the green pigment involved.

Based on above context, explain photosynthesis in simple terms.""",

    "Detailed Context": """Photosynthesis is a process used by plants, algae, and some bacteria.
It converts light energy, usually from the sun, into chemical energy stored in glucose.
The process occurs in the chloroplasts of plant cells.
Chlorophyll absorbs light, mainly in the blue and red wavelengths.
Water molecules are split during this process, releasing oxygen as a byproduct.
Carbon dioxide from the air is also used in this process.
The overall equation is: 6CO2 + 6H2O + light energy → C6H12O6 + 6O2

Based on the detailed context above, give a comprehensive explanation of photosynthesis."""
}

for context_type, prompt in prompts.items():
    print(f"\n CONTEXT TYPE: {context_type}")
    print("-" * 40)
    response = ask_llm(prompt)
    print(response)
    print("=" * 60)
    time.sleep(2)

print("\nExperiment 3 Complete!")
```

```python
# EXPERIMENT 4: Detecting LLM Hallucinations

print("EXPERIMENT 4: HALLUCINATION ANALYSIS")
print("=" * 60)

hallucination_prompts = {
    "Fake Person": "Tell me about Dr. James Wellington who won the Nobel Prize in AI in 2023.",

    "Fake Event": "Explain what happened at the Global AI Summit held in Mars in 2022.",

    "Ambiguous Fact": "What is the exact population of the moon colony as of 2024?",

    "Real Fact Check": "Who created the Python programming language and when?"
}

for test_name, prompt in hallucination_prompts.items():
    print(f"\n TEST: {test_name}")
    print("-" * 40)
    print(f"PROMPT: {prompt}")
    print("\nLLM RESPONSE:")
    response = ask_llm(prompt, temperature=0.1)
    print(response)
    print("=" * 60)
    time.sleep(2)

print("\nExperiment 4 Complete!")
```
